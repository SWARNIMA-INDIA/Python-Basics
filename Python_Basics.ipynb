#Python Basics








1. What is Python, and why is it popular?
  - Python is a high-level, interpreted programming language known for its     simplicity, readability, and wide range of libraries.
  It is popular because it is easy to learn and used in many fields like data science, AI, and web development.

2. What is an interpreter in Python?
  - Interpreter reads the code line by line and runs it.

3. What are pre-defined keywords in Python?
  - Keywords are reserved words that have special meaning in Python.
  (e.g., if, else, for, while, True, False, def).

4. Can keywords be used as variable names?
  - No, we cannot use keywords as variable names.

5. What is mutability in Python?
  - Mutability means we can change the value after creation.
  (e.g., lists are mutable; tuples and strings are immutable.)

6. Why are lists mutable, but tuples are immutable?
  - Lists can be modified because they are designed to change frequently, while tuples are fixed and used for constant data.

7. What is the difference between “==” and “is” operators in Python?
   - “==” checks if two values are equal or not.
   - “is” checks if both names or variables are pointing to the same object in memory.

8. What are logical operators in Python?
   - There are three logical operators in python:
     - AND - Returns True if both operands are true.
     - OR - Returns True if at least one operand is true.
     - NOT -  Flips the Boolean value of its operand; returns True if the operand is false.

9. What is type casting in Python?
   - Changing one data type to another is called type casting.
   Example: int("5") changes string to integer.
            float(3) changes integer to float.

10. What is the difference between implicit and explicit type casting?
   - Implicit: Python changes type automatically.
   - Explicit: We change type using functions like int(), float(), etc.

11. What is the purpose of conditional statements in Python?
   - Conditional statements check conditions and run code accordingly.
   Example: if, elif, else.
   - key purposes are:
     - Decision making — like turning left or right.
     - Control program flow to handle different scenarios (such as user input, data values, or states).
     - Enable algorithms to perform varies actions depending on context or requirements.

12. How does the elif statement work?
   -  It checks additional conditions after if. If (if) is false, Python checks elif.

13. What is the difference between for and while loops?
   - For loop: Use it when we know how many times to repeat (like going through a list).
   - While loop: Use it when we keep repeating until a condition changes (we do not know how many times).






```python
#  14. Describe a scenario where a while loop is more suitable than a for loop.
''' - A scenario where a while loop is more suitable than a for loop is when building a password.
     Here it continues until the user inputs the correct password. We cannot predict how many
     tries the user will need, making a while loop perfect for this case.'''

password = ""
while password != "python":
  password = input("Enter password:")
print("Access Granted!")
```

```python
#(5). Create a list and tuple in Python, and demonstrate how attempting to change an element works differently for each.

# A list can be changed (mutable)
my_list = [1, 2, 3]
print("Original list:", my_list)
my_list[0] = 10
print("After changing list:", my_list)

# A tuple cannot be changed (immutable)
my_tuple = (1, 2, 3)
print("Original tuple:", my_tuple)
try:
    my_tuple[0] = 10   # This will cause an error
except TypeError:
    print("Tuples cannot be changed once created.")
```

```python
# (6).  Write a function to demonstrate the behavior of mutable and immutable arguments.
# Mutable example (list)
def change_list(lst):
    lst.append(4)

# Immutable example (number)
def change_number(num):
    num += 5
    return num

# Test
numbers = [1, 2, 3]
value = 10

change_list(numbers)
print("List after function call:", numbers)

new_value = change_number(value)
print("Number after function call:", value)
print("Returned number:", new_value)
```

```python
#(9). Write a Python program to convert user input from string to integer, float, and boolean types.
s = input("Enter something: ")

# Convert to integer or float only if it's a number
if s.isdigit():  # Checks if all characters are digits (0–9)
    print("As integer:", int(s))
    print("As float:", float(s))
else:
    print("Cannot convert to number because it's not numeric.")

# Convert to boolean (this always works)
print("As boolean:", bool(s))
```

```python
#(10). Write code to demonstrate type casting with list elements.
# Original list (all elements are strings)
numbers = ["10", "20", "30"]
print("Before type casting:", numbers)
print("Data type of first element:", type(numbers[0]))

# Convert (type cast) all elements from string to integer
numbers = [int(x) for x in numbers]

# Show result after conversion
print("After type casting:", numbers)
print("Data type of first element:", type(numbers[0]))
```

```python
#(11). Write a program that checks if a number is positive, negative, or zero.
# Ask the user to enter a number
num = float(input("Enter any number: "))

# Check if the number is positive
if num > 0:
    print("The number is positive.")

# Check if the number is negative
elif num < 0:
    print("The number is negative.")

# Otherwise, it must be zero
else:
    print("The number is zero.")
```

```python
#(12). Write a for loop to print numbers from 1 to 10.
i = 1
while i < 11:
    print(i)
    i += 1
```

```python
#(13). Write a Python program to find the sum of all even numbers between 1 and 50.
total = 0
for i in range(1, 51):
    if i % 2 == 0:
        total += i
print(total)
```

```python
# (14). Write a program to reverse a string using a while loop.
text = input("Enter a string: ")
reversed_text = ""
i = len(text) - 1

while i >= 0:
    reversed_text += text[i]
    i -= 1

print("Reversed string:", reversed_text)
```

```python
# (15). Write a Python program to calculate the factorial of a number provided by the user using a while loop.
num = int(input("Enter a number: "))
fact = 1
i = 1

while i <= num:
    fact *= i
    i += 1

print("Factorial of", num, "is", fact)
```
